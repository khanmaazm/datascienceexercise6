```python
#Base Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Data Science Modelling Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
advertisingdf = pd.read_csv('advertising.csv')
advertisingdf
```

```python
advertisingdf.drop(['Ad Topic Line','City','Country','Timestamp'], axis = 1, inplace = True)
```

```python
advertisingdf
```

```python
advertisingdf.columns = ['DailyTimeSpentonSite', 'Age', 'AreaIncome', 
       'DailyInternetUsage', 'Male', 'ClickedonAd']
```

```python
advertisingdf
```

```python
X = advertisingdf.drop('ClickedonAd', axis = 1)
X
```

```python
Y = advertisingdf['ClickedonAd']
Y
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size = 0.25)
```

```python
y_test
```

```python
mylogisticmodelvariable = LogisticRegression()
mylogisticmodelvariable
```

```python
mylogisticmodelvariable.fit(X_train, Y_train)
```

```python
mylogisticmodelvariable
```

```python
mydecisiontreemodel = DecisionTreeClassifier()
mydecisiontreemodel
```

```python
mydecisiontreemodel.fit(X_train,Y_train)
```

```python
y_pred = mylogisticmodelvariable.predict(x_test)
y_pred
```

```python
y_preddectree = mydecisiontreemodel.predict(x_test)
y_preddectree
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
```

```python
rfcmodel = RandomForestClassifier(n_estimators=50)
rfcmodel
```

```python
rfcmodel.fit(X_train, Y_train)
```

```python
y_predrandomforest = rfcmodel.predict(x_test)
y_predrandomforest
```

```python
clickedonpreds = pd.DataFrame()
```

```python
clickedonpreds['YTest'] = y_test
clickedonpreds['YPredLogistic'] = y_pred
clickedonpreds['YPredDecTree'] = y_preddectree
clickedonpreds['YPredRandomForest'] = y_predrandomforest
clickedonpreds
```

```python
print('No of Test prediction Errors in Logistic ' , clickedonpreds[(clickedonpreds['YTest'] != clickedonpreds['YPredLogistic'])].count()['YTest'])
print('No of Test prediction Errors in Decision tree ' , clickedonpreds[(clickedonpreds['YTest'] != clickedonpreds['YPredDecTree'])].count()['YTest'])
print('No of Test prediction Errors in Random Forest ' , clickedonpreds[(clickedonpreds['YTest'] != clickedonpreds['YPredRandomForest'])].count()['YTest'])
```

```python
print('No of Test prediction correct in Logistic ' , clickedonpreds[(clickedonpreds['YTest'] == clickedonpreds['YPredLogistic'])].count()['YTest'])
print('No of Test prediction correct in Decision tree ' , clickedonpreds[(clickedonpreds['YTest'] == clickedonpreds['YPredDecTree'])].count()['YTest'])
print('No of Test prediction correct in Random Forest ' , clickedonpreds[(clickedonpreds['YTest'] == clickedonpreds['YPredRandomForest'])].count()['YTest'])
```
